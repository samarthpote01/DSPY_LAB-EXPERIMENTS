**Q.1 Explain input and output characteristics for f_oneway function from stats library.**

**Ans:-**

**Input Characteristics:**

* **Samples:** Takes two or more 1D array-like samples (groups) as input.

* **Dimensions:** Accepts multidimensional input arrays. If inputs are 2D, the test is calculated along the first axis (columns) by default.

* **Sizes:** Sample sizes can differ (unbalanced designs).

* **Axis:** Optional axis parameter defines the input axis along which to compute the statistic (default is 0).

* **NaN Handling:** The nan_policy parameter (propagate, omit, raise) dictates how to handle missing values.
Method: By default, it performs a standard one-way ANOVA. In recent versions, equal_var=False can perform Welch’s ANOVA.

**Output Characteristics:**

* **Return Type:** Returns a tuple → (F-statistic, p-value)

* **F-Statistic:** A float measuring the ratio of variance between groups relative to variance within groups.

* **P-Value:** A float indicating the probability that the observed group means could occur under the null hypothesis.

**Edge Cases:**

* Returns (NaN, NaN) if input arrays have zero length or all inputs are of size 1.

* Returns (∞, 0) if all values within each group are identical but differ across groups.

* Emits warnings like ConstantInputWarning or RuntimeWarning in such cases.

**Q.2 Explain results of sm.stats.anova_lm fucntion.**

**Ans:-**

The anova_lm() function is available in the statsmodels library. It is used to perform ANOVA using a linear model and provides results in table format.

It helps determine whether the mean values of groups are significantly different.

**Syntax:**

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
import pandas as pd
import numpy as np

# Create a sample DataFrame for demonstration
data = pd.DataFrame({
    'y': [10, 12, 15, 11, 13, 20, 22, 25, 21, 23],
    'x': ['A', 'A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'B']
})

model = ols('y ~ x', data=data).fit()
sm.stats.anova_lm(model)
```

**Output of anova_lm():**

The function returns a table with the following columns:

* **sum_sq** → Sum of squares (variation due to each factor)

* **df** → Degrees of freedom

* **F** → F-statistic value

* **PR(>F)** → p-value

**Interpretation:**

* If **p-value < 0.05**, reject the null hypothesis → significant difference exists

* If **p-value ≥ 0.05**, fail to reject → no significant difference

**Q.3 What role ANOVA plays in data analytics?**

**Ans:-**

ANOVA (Analysis of Variance) plays an important role in data analytics:

* Helps compare means of multiple groups simultaneously

* Reduces error compared to multiple t-tests

* Identifies whether variation is due to actual differences or randomness

* Used in business, healthcare, manufacturing, and research

Example: Comparing sales performance across different regions

**Q.4 When to use one way ANOVA? Explain with real world case.**

**Ans:-**

**When to use:**

* When comparing one independent variable with multiple groups

* When checking if group means are significantly different

**Real-world example:**

* A teacher wants to check whether students from 3 different classes have different average marks.

* Null hypothesis: All class means are equal
Alternative hypothesis: At least one class mean differs

If p-value < 0.05 → performance differs significantly

**Q.5 When to use 2 way ANOVA ? Explain with real world case.**

**Ans:-**

**When to use:**

* To analyze both individual and interaction effects

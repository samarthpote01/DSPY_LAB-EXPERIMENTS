```python
import pandas as pd

# IMPORTANT: Replace 'path/to/your/dataset.csv' with the actual path to your dataset file.
# Ensure your CSV file has 'text' and 'label' columns for the model.

# For now, let's create a dummy DataFrame to avoid FileNotFoundError and proceed with model building.
data = {'text': ['This is a positive review', 'This is a negative comment', 'Great product', 'Terrible service', 'Highly recommended'],
        'label': [1, 0, 1, 0, 1]}
df = pd.DataFrame(data)

# To use your own data, uncomment the line below and provide the correct path:
# df = pd.read_csv('path/to/your/dataset.csv')

# Display the first few rows of the DataFrame to verify it loaded correctly
print("DataFrame loaded successfully. First 5 rows:")
display(df.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

X = df['text']
y = df['label']

# Convert text
cv = CountVectorizer()
X = cv.fit_transform(X)

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y)

# Train
model = MultinomialNB()
model.fit(X_train, y_train)

# Accuracy
model.score(X_test, y_test)
```

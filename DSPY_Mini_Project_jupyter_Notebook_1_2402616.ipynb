**Aim :** To preprocess the LDC2025T10 dataset by selecting relevant data, cleaning inconsistencies, and transforming it into a structured format.

**Notebook-1**

**Dataset Description**

The dataset contains:

* Textual data (sentences/documents)
* Linguistic annotations
* Metadata

**Data Selection**

Selected attributes:

* text → main content
* label → category/output

**Reason:**
* These are required for ML modeling
* Removes unnecessary metadata

**Code:-**

```python
import pandas as pd
import json
import os # Import os module to check for file existence

# Placeholder for data.json creation if it doesn't exist
# In a real scenario, you would upload your 'data.json' file to your Colab environment
# or ensure the correct path is provided.
if not os.path.exists('data.json'):
    print("data.json not found. Creating a dummy data.json for demonstration.")
    dummy_data = [
        {"text": "This is a sample sentence.", "label": "positive"},
        {"text": "Another example here.", "label": "negative"}
    ]
    with open('data.json', 'w') as f:
        json.dump(dummy_data, f)

# Load dataset
with open('data.json', 'r') as f:
    data = json.load(f)

df = pd.json_normalize(data)

# Select important columns
df = df[['text', 'label']]

df.head()
```

**Data Cleaning:-**

```python
df.isnull().sum()
df.dropna(inplace=True)

df['text'] = df['text'].str.lower()
df['text'] = df['text'].str.replace('[^a-zA-Z]', ' ', regex=True)
```

**Data Wrangling:-**

```python
df['text_length'] = df['text'].apply(len)
```

**Attribute Description:-**

| Attribute | Description | Type |
| --- | --- | --- |
| text | sentence data | object |
| label | category | object |
| text_length | length of sentence | int |

**Conclusion:-**

Dataset is cleaned and ready for EDA.

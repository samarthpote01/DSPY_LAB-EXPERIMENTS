# NON-GRAPHICAL ANALYSIS

**1. Sentence Length Analysis (Univariate)**

```python
import pandas as pd

# Example data for demonstration
data = {'text': ['This is a sample sentence.', 'Another sentence here.', 'Short text.']}
df = pd.DataFrame(data)

# Calculate text length
df['text_length'] = df['text'].apply(len)

df['text_length'].describe()
```

**Objective:** Understand sentence size

**Observation:** Moderate length

**2. Label Distribution (Multivariate)**

```python
import pandas as pd

# Ensure df is defined with a 'label' column for this example
# (Assuming df from previous cells, adding a 'label' column for demonstration)

# If df is not yet defined (e.g., if you run this cell independently),
# uncomment and run the following lines to create a sample df:
# data = {'text': ['This is a sample sentence.', 'Another sentence here.', 'Short text.']}
# df = pd.DataFrame(data)
# df['text_length'] = df['text'].apply(len)

# Adding a sample 'label' column for demonstration purposes
# In a real scenario, you would load or generate your actual labels.
df['label'] = ['positive', 'negative', 'positive'] # Sample labels

df['label'].value_counts()
```

**Objective:** Check class balance

**Observation:** Some labels dominate

# GRAPHICAL ANALYSIS
**3. Histogram**

```python
import matplotlib.pyplot as plt
plt.hist(df['text_length'])
plt.show()
```

**4. Bar Chart**

```python
df['label'].value_counts().plot(kind='bar')
plt.show()
```
